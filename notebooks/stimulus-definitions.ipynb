# Mathematical Definitions Of The Color-Cue Paradigm

This notebook mirrors the standalone LaTeX document `stimulus-definitions.tex`. It summarizes the old stimulus generator and the current two-stage per-stimulus noise calibration paradigm.

## Old Stimulus Generator

A stimulus is generated on an image lattice `(u, v)` from a complex spatial-noise field plus a deterministic signal vector:

```text
X(u, v) = N(u, v) + kappa * exp(i theta)
```

The displayed pixel value is the angle of this complex image, remapped through the custom colormap. The old texture noise can average away under circular pooling, motivating direct manipulation of the latent mean hue angle `theta`. Going forward, `kappa` is fixed to 0.1.

## 2AFC Trial Structure

Each trial uses a left-right pair centered on `theta0`:

```text
theta_L* = theta0 - Delta theta / 2
theta_R* = theta0 + Delta theta / 2
```

External noise is injected into the latent theta values before rendering:

```text
theta_L = theta_L* + epsilon_L,ext
theta_R = theta_R* + epsilon_R,ext
```

The current main design uses independent external noise on the two sides. Shared external noise remains useful as a control for common-mode effects.

## Per-Stimulus Calibration Model

The revised observer model defines noise per stimulus:

```text
theta_hat_L = theta_L + epsilon_L,int + epsilon_L,ext
theta_hat_R = theta_R + epsilon_R,int + epsilon_R,ext
```

With independent left/right noise, the comparison variable has variance:

```text
Var(theta_hat_R - theta_hat_L) = 2 * (sigma_int^2 + sigma_ext^2)
```

The factor of two belongs to the comparison. The calibration quantities `sigma_int`, `sigma_ext`, and `sigma_target` are per-stimulus standard deviations.

## Experiment 1: Baseline Internal Noise

With no external noise, the Gaussian 2AFC model is:

```text
P(choose right | Delta theta) = Phi(Delta theta / (sqrt(2) * sigma_int))
```

For criterion performance `p`, the baseline threshold satisfies:

```text
JND_p = sqrt(2) * sigma_int * Phi^{-1}(p)
```

So:

```text
sigma_int = JND_p / (sqrt(2) * Phi^{-1}(p))
```

## Experiment 2: Matched Total Noise

The experimenter chooses target total per-stimulus noise levels. For each subject:

```text
sigma_target^2 = sigma_int^2 + sigma_ext^2
sigma_ext = sqrt(sigma_target^2 - sigma_int^2)
```

This is feasible only when `sigma_target >= sigma_int`. If two subjects have different baseline thresholds but receive subject-specific `sigma_ext` values that produce the same `sigma_target`, their psychometric curves should overlap in the matched-noise task.

## Correlated Internal Noise Caveat

If left and right internal noises are correlated with correlation `rho_int`, then:

```text
Var(epsilon_R,int - epsilon_L,int) = 2 * sigma_int^2 * (1 - rho_int)
```

A standard 2AFC curve identifies the effective differential noise:

```text
sigma_int,diff = sigma_int * sqrt(1 - rho_int)
```

not `sigma_int` and `rho_int` separately. The working model is `rho_int = 0`; if shared-noise controls are not flat, report the calibration as an effective differential-noise calibration.
